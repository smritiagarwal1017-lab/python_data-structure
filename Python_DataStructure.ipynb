**Quest 1**-  What are data structures, and why are they important?

**Ans 1**- Data structures are specialized formats or organizing, storing, and managing data in a computer's memory so that it can be used efficiently. The importance iof data structure in python are:

*   Efficiency and Performance: Using the correct data structure can significantly improve the speed and performance of a program.

*   Code Organization and Readability: Data structures provide a logical way to organize data, making the code cleaner, more maintainable, and easier for other developers to understand and debug.

* Problem-Solving: Different data structures are designed to solve specific types of problems efficiently.





**Quest 2**-  Explain the difference between mutable and immutable data types with examples?

**Ans 2**- **Mutable data types** can be modified after they are created, directly changing the original object in memory where as **Immutable data types**, once created, cannot be altered; any operation that seems to modify them actually results in a new object being created.

Example for mutable data type are list, dictionary, sets where as example for immutable data type are string, tuples and frozensets.

**Quest 3**- What are the main differences between lists and tuples in Python?

**Ans 3**- The difference between Python lists and tuples is that **lists** are mutable (changeable) while **tuples** are immutable (unchangeable). This fundamental distinction impacts their syntax, performance, and typical use cases. **Lists** cannot be used in Dictionary keys where **Tuples** can be used in Dictionary keys.

**Quest 4**- Describe how dictionaries store data?

**Ans 4**- Dictionaries store data as key-value pairs using an underlying data structure known as a hash table (or hash map).

**Quest 5**- Why might you use a set instead of a list in Python?

**Ans 5**- You might use a Python set instead of a list when you need to store only unique elements, require fast membership testing (checking if an item is present), or need to perform mathematical set operations like unions and intersections.

**Quest 6**-  What is a string in Python, and how is it different from a list?

**Ans 6**- In Python, a string is an immutable sequence of Unicode characters used to represent text, while a list is a mutable ordered collection of items of potentially different data types. The primary difference lies in mutability: lists can be changed after creation, while strings cannot.

**Quest 7**- How do tuples ensure data integrity in Python?

**Ans 7**- Tuples ensure data integrity in Python primarily through their core characteristic of immutability. This means that once a tuple is created, its elements cannot be changed, added, or removed, which guarantees that the data remains constant and protected from accidental modification throughout the program's execution.

**Quest 8**- What is a hash table, and how does it relate to dictionaries in Python?

**Ans 8**- A hash table is a data structure used to store and retrieve data in key-value pairs very quickly, typically with an average time complexity of O(1). The Python 3 dictionary (dict) is a built-in implementation of a hash table.

**Quest 9**- Can lists contain different data types in Python?

**Ans 9**- Yes, Python lists can contain elements of different data types within the same list. This flexibility makes them a versatile and powerful data structure.

**Quest 10**-  Explain why strings are immutable in Python?

**Ans 10**- Strings are immutable in Python, meaning their value cannot be changed after creation, as a deliberate design choice to provide benefits such as thread safety, predictable behavior, and efficient memory management.

**Quest 11**- What advantages do dictionaries offer over lists for certain tasks?

**Ans 11**- Dictionaries offer superior performance for searching, inserting, and deleting data compared to lists, providing average-case
 time complexity due to their hash table structure. They excel at storing data as key-value pairs, allowing for fast, label-based retrieval rather than relying on slow numerical index lookups or scanning through large datasets.

**Quest 12**-  Describe a scenario where using a tuple would be preferable over a list?

**Ans 12**- A scenario where using a tuple is preferable over a list is when you need to store related, heterogeneous data that should remain constant throughout the program's execution, such as a database record or a set of coordinates.

**Quest 13**- How do sets handle duplicate values in Python?

**Ans 13**- In Python, sets automatically remove duplicate values; each element in a set is unique. This behavior is a core characteristic of the set data type, designed for operations that require a collection of distinct items.

**Quest 14**- How does the “in” keyword work differently for lists and dictionaries?

**Ans 14**- The in keyword in Python is a membership operator that checks for the presence of an item in a collection.

In **Lists** the operator checks if the specified value is present as an element in the list. It performs a value search across the list items where as in **Dictionaries** the operator checks for the presence of a key in the dictionary, not a value.

**Quest 15**- Can you modify the elements of a tuple? Explain why or why not?

**Ans 15**- No, you cannot modify the elements of a tuple after it has been created because tuples are immutable data structures in Python. Once a tuple is defined, its size and content are fixed. You cannot add, remove, or replace individual items using indexing or slicing assignment.This immutability ensures data integrity, which is useful for storing data that should remain constant throughout the program, such as configuration settings or database records.

**Quest 17**- Describe the time complexity of accessing elements in a dictionary?

**Ans 17**- Accessing elements in a dictionary has an average-case time complexity of O(1) (constant time) and a worst-case complexity of O(n) (linear time).

**Quest 18**- In what situations are lists preferred over dictionaries?

**Ans 18**- Lists are preferred over dictionaries when data requires a specific, ordered sequence, allows duplicate values, or when fast index-based access (
) is needed rather than key-based lookup. Lists are ideal for ordered collections, such as stacks or queues, and for scenarios requiring frequent iteration, sorting, or slicing.

**Quest 19**- Why are dictionaries considered unordered, and how does that affect data retrieval?

**Ans 19**- Dictionaries are considered unordered because they map keys to values using hash tables for rapid, O(1) retrieval, rather than storing elements in a sequential, indexable order.

**Quest 20**- Explain the difference between a list and a dictionary in terms of data retrieval?

**Ans 20**- The primary difference in data retrieval is that a **list** uses sequential integer indices for access, while a **dictionary** uses unique, descriptive keys. **Dictionaries** are optimized for very fast, direct lookups by key, whereas searching a **list** for a specific value (not by index) is slower for large collections.
